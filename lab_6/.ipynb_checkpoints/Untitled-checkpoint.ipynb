```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import datetime as dt

import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
```

```python
retail_df = pd.read_csv("Online_Retail.csv", sep=",", encoding="ISO-8859-1", header=0)
retail_df.head()
```

```python
# missing values
round(100*(retail_df.isnull().sum())/len(retail_df), 2)
```

```python
# drop all rows having missing values
retail_df = retail_df.dropna()
retail_df.shape
```

```python
retail_df.head()
```

```python
# new column: amount 
retail_df['amount'] = retail_df['Quantity']*retail_df['UnitPrice']
retail_df.head()
```

```python
#R (Recency): Number of days since last purchase
#F (Frequency): Number of tracsactions
#M (Monetary): Total amount of transactions (revenue contributed)

# monetary
grouped_df = retail_df.groupby('CustomerID')['amount'].sum().reset_index()
# grouped_df = grouped_df.reset_index()
grouped_df.head()
```

```python
# frequency
frequency = retail_df.groupby('CustomerID')['InvoiceNo'].count().reset_index()
# frequency = frequency.reset_index()
frequency.columns = ['CustomerID', 'frequency']
frequency.head()
```

```python
# Merge with custom suffixes
grouped_df = pd.merge(grouped_df, frequency, on='CustomerID', how='inner', suffixes=('', '_drop'))

# Drop the extra column if it exists
grouped_df = grouped_df.drop(columns=[col for col in grouped_df.columns if col.endswith('_drop')], errors='ignore')

grouped_df.head()
```

```python
retail_df.head()
```

```python
# recency
# convert to datetime
retail_df['InvoiceDate'] = pd.to_datetime(retail_df['InvoiceDate'], 
                                          format='%d-%m-%Y %H:%M')

retail_df.head()
```

```python
# compute the max date
max_date = max(retail_df['InvoiceDate'])
max_date
```

```python
# compute the diff
retail_df['diff'] = max_date - retail_df['InvoiceDate']
retail_df.head()
```

```python
# recency
last_purchase = retail_df.groupby('CustomerID')['diff'].min()
last_purchase = last_purchase.reset_index()
last_purchase.head()
```

```python
# merge
grouped_df = pd.merge(grouped_df, last_purchase, on='CustomerID', how='inner')
grouped_df.columns = ['CustomerID', 'amount', 'frequency', 'recency']
grouped_df.head()
```

```python
# number of days only
grouped_df['recency'] = grouped_df['recency'].dt.days
grouped_df.head()
```

```python
# 1. outlier treatment
plt.boxplot(grouped_df['recency'])
```

# two types of outliers:
# - statistical
# - domain specific

```python
# removing (statistical) outliers
Q1 = grouped_df.amount.quantile(0.05)
Q3 = grouped_df.amount.quantile(0.95)
IQR = Q3 - Q1
grouped_df = grouped_df[(grouped_df.amount >= Q1 - 1.5*IQR) & (grouped_df.amount <= Q3 + 1.5*IQR)]
```

```python
# outlier treatment for recency
Q1 = grouped_df.recency.quantile(0.05)
Q3 = grouped_df.recency.quantile(0.95)
IQR = Q3 - Q1
grouped_df = grouped_df[(grouped_df.recency >= Q1 - 1.5*IQR) & (grouped_df.recency <= Q3 + 1.5*IQR)]
```

```python
# outlier treatment for frequency
Q1 = grouped_df.frequency.quantile(0.05)
Q3 = grouped_df.frequency.quantile(0.95)
IQR = Q3 - Q1
grouped_df = grouped_df[(grouped_df.frequency >= Q1 - 1.5*IQR) & (grouped_df.frequency <= Q3 + 1.5*IQR)]
```

```python
# 2. rescaling
rfm_df = grouped_df[['amount', 'frequency', 'recency']]
```

```python
# instantiate
scaler = StandardScaler()
```

```python
# fit_transform
rfm_df_scaled = scaler.fit_transform(rfm_df)
rfm_df_scaled.shape
```

```python
rfm_df_scaled = pd.DataFrame(rfm_df_scaled)
rfm_df_scaled.columns = ['amount', 'frequency', 'recency']
rfm_df_scaled.head()
```

```python
#4. Modelling
# k-means with some arbitrary k
kmeans = KMeans(n_clusters=4, max_iter=50)
kmeans.fit(rfm_df_scaled)

kmeans.labels_
```

```python
# assign the label
grouped_df['cluster_id'] = kmeans.labels_
grouped_df.head()
```

```python
# plot
sns.boxplot(x='cluster_id', y='amount', data=grouped_df)
```
